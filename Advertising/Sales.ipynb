Import all the required python libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Importing the data file 'advertising.csv ' by using python pandas library

```python
# Loading the dataset
df = pd.read_csv("advertising.csv")
```

## Data Analytics 

Plotting a graph for the "TV" and "Sales" 

```python
sns.scatterplot(data = df,x='TV',y='Sales')
plt.show()
```

From above visual we can see that as we increase the advertising cost in tv the sales will increase

Plotting a graph which describes the relation of "Radio" and "Sales"

```python
sns.scatterplot(data = df,x='Radio',y='Sales')
plt.show()
```

From above visual we can understand that as advertising cost of "Radio" increases the sales may increase or may not increase
as the data is more dispersed form the intital stage only

Plotting a graph for "Newspaper" and "Sales"

```python
sns.scatterplot(data= df,x='Newspaper',y='Sales')
plt.show()
```

From above visual we can understand that if advertising cost increase in Newspaper the Sales may not increase as data point are far away from each other as the advertising cost increases

Plotting a pairgraph for better understanding of these three visuals and compare them

```python
sns.pairplot(data = df,x_vars=['TV','Radio','Newspaper'],y_vars='Sales')
plt.show()
```

Now we know that if we focus on advertising in TV we can get maximum sales from that 

Plotting a pairplot along with the regression line in the graph 

```python
sns.pairplot(data = df,x_vars=['TV','Radio','Newspaper'],y_vars=['Sales'],kind ='reg')
plt.show()
```

Plotting a heat map correlation with each other

```python
sns.heatmap(df.corr(),annot=True,fmt='.2f')
```

Form above all analysis we understood that if we advertise in TV we can get in high Sales

Selecting the column from the data for the prediction

```python
x=df[['TV']]
y=df[['Sales']]
```

Spliting the data in training and testing data with the help of train_test_split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Using the LinearRegression model for training and prediction

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

Predicting the Train data with Linear Regression

```python
x_train_predict = model.predict(x_train)
```

```python
print(x_train_predict)
```

Predicting the Test data with Linear Regression

```python
x_test_predict = model.predict(x_test)
```

```python
print(x_test_predict)
```

Plotting a graph that visualize the original train data and predicted trian data 

```python
plt.scatter(y_train,x_train_predict)
plt.title("Original VS Predicted Values")
plt.xlabel("Original Value")
plt.ylabel("Predicted Value")
plt.show()
```

From above visuval we can see that as original train value increases the predicted train value also increases

Plotting a graph that visualize the original test data and predicted test data 

```python
plt.scatter(y_test,x_test_predict)
plt.title("Original VS Predicted Values")
plt.show()
```

From above visuval we can see that as original train value increases the predicted train value also increases

Comparing Original VS Predicted Values in both Train and Test data

```python
plt.suptitle("Train VS Test")
plt.subplot(1,2,1)
plt.scatter(y_train,x_train_predict)
plt.title("Train")
plt.subplot(1,2,2)
plt.scatter(y_test,x_test_predict)
plt.title("Test")
plt.show()
```
